## Скелетное представление изображения

Работу выполнил Дмитрий Лобузнов

Скелет изображения - множество точек, равноудаленных от границ фигуры.
Для нахождения скелета изображения воспользуемся библиотекой skimage. В этой библиотеке есть реализация алгоритма нахождения скелета изображения.

Ниже приведен пример кода для следующих изображений
- сулуэта лошади
- синтетическое изображение( рандомные шары ) 


```python
from skimage.morphology import skeletonize
from skimage import data
import matplotlib.pyplot as plt
from skimage.util import invert
import numpy as np
```

### Пример 1. Силуэт лошади 

```python
# загружаем готовое изображение из библиотеки.
image = invert(data.horse())
# img = data.cat()

# находим скелет
skeleton = skeletonize(image)

# визуализация изображений
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4),
                         sharex=True, sharey=True)
ax = axes.ravel()


ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Входное изображение', fontsize=15)

ax[1].imshow(skeleton, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('скелетное представление', fontsize=15)

fig.tight_layout()
plt.show()
```

### Пример 2. Шары 

```python

blobs = data.binary_blobs(500, blob_size_fraction=.1,
                          volume_fraction=.35, seed=42)

skeleton = skeletonize(blobs)
skeleton_lee = skeletonize(blobs, method='lee')

fig, axes = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(blobs, cmap=plt.cm.gray)
ax[0].set_title('Входное изображение')
ax[0].axis('off')

ax[1].imshow(skeleton, cmap=plt.cm.gray)
ax[1].set_title('Скелетное представление')
ax[1].axis('off')


fig.tight_layout()
plt.show()
```
